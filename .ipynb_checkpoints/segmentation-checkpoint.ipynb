```python
#Your code here:

from sklearn.cluster import KMeans, DBSCAN
from sklearn.neighbors import NearestNeighbors
```

```python
K = range(1, 15)
sum_squared_distances = []

for k in K:
    model = KMeans(n_clusters=k).fit(df_pca)
    sum_squared_distances.append(model.inertia_)

plt.plot(K, sum_squared_distances)
plt.xlabel('K Values')
plt.ylabel('Sum of Squared Distances')
plt.title('Elbow Method')

plt.show()
```

```python
# A partir de la grafica del codo podemos ver con 6 clusters podemos agrupar adecuadamente los datos
df_labels = df_pca.copy()

kmeans = KMeans(n_clusters=6, random_state=42)
kmeans.fit(df_pca)
df_labels['labels'] = kmeans.labels_
```

```python
df_labels['labels'].value_counts()
```

```python
df['labels'] = kmeans.labels_
df.groupby('labels').mean()
```

```python
# Podemos ver que esta clasificando por lo vemos correctamente un grupo. Ya que como vemos el grupo 2 tiene una edad 
# media de 101 años. Este grupo corresponde al de los outliers con una edad muy avanzada.
```

```python
fig, axes = plt.subplots(13,6, figsize=(25,50))
combs=[(i,j) for j in range(13) for i in range(13) if i!=j]
for ax in axes.flat:
    comb=combs.pop()
    ax.scatter(df.iloc[:,comb[0]], df.iloc[:,comb[1]], c=df.labels,cmap='nipy_spectral_r')
    ax.set_xlabel(df.columns[comb[0]])
    ax.set_ylabel(df.columns[comb[1]])
plt.tight_layout()
```
