```python
#Your code here:

#Category Encoding

from sklearn.preprocessing import LabelEncoder
data_2 = data_1.copy()
```

```python
le = LabelEncoder()
data_2['gender'] = le.fit_transform(data_2['gender'])
```

```python
data_2=pd.get_dummies(data_2,columns=['gradyear'],drop_first=True)
```

```python
data_2
```

```python
data_append=data_2.loc[:,['gender','gradyear_2007','gradyear_2008','gradyear_2009']]

data_append
```

```python
#Outliers & Standardization

from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
```

```python
data_3=data_2[data_2.columns[1:-3]]
for i in data_2.columns[1:-3]:
    data_3[i] = scale.fit_transform(data_2[i].values.reshape(-1,1))
data_3
```

```python
data_3
```

```python
data_4=pd.merge(data_3, data_append ,how="left", on=data_3.index)
data_4
```

```python
data_4.drop(columns=['key_0'],inplace=True)
```

```python
from sklearn.decomposition import PCA
pca=PCA().fit(data_4)
plt.figure(figsize=(15,5))
plt.plot(pca.explained_variance_ratio_.cumsum(),marker='o',linestyle='--')
```

```python
pca=PCA(n_components=27)
principalComponents=pca.fit_transform(data_4)
data_5=pd.DataFrame(data=principalComponents)
data_5.head()
```

## 4. Modeling

Now that your data is ready, apply one (or more) unsupervised learning model to find the groups of teens that exist:
* What model did you choose?
* What is the optimal number of clusters?
* Once your clusters are defined, how do you interpret them? - What type of teens there are? What are the differences between them?

```python
#Your code here:
K = range(1, 15)
sum_squared_distances = []
for k in K:
    model = KMeans(n_clusters=k).fit(data_5)
    sum_squared_distances.append(model.inertia_)
plt.figure(figsize=(15, 5))
plt.plot(K, sum_squared_distances)
plt.xlabel('K Values')
plt.ylabel('Sum of Squared Distances')
plt.title('Elbow Method')
plt.show()

```

```python
kmeans = KMeans(n_clusters=6, random_state=42)
teens_clusters = kmeans.fit_predict(data_5)
data_1['labels']=teens_clusters
data_1['labels'].value_counts()
```

```python
data_1.groupby('labels').mean().T
```
